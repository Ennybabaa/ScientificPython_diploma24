#### **Exercise 2:**

Integrate the function `x**2 + 10*np.sin(x)` on the interval `[-10,10]`.
Find a way to make a plot like in textbooks where you color the area under the curve that you integrate.
Now integrate the same function symbolically using `sympy`, then put borders into it and compare the results.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
import sympy as sp

x = sp.symbols('x')
f = x**2 + 10*sp.sin(x)
func = lambda x: x**2 + 10*np.sin(x)

result_numeric, _ = quad(func, -10, 10)
result_symbolic = sp.integrate(f, (x, -10, 10))

x_values = np.linspace(-10, 10, 1000)
y_values = func(x_values)

fig, ax = plt.subplots()
ax.plot(x_values, y_values, 'b', linewidth=2, label='Function')

x_fill = np.linspace(-10, 10, 1000)
y_fill = func(x_fill)
ax.fill_between(x_fill, y_fill, color='lightblue', alpha=0.5, label='Area under the curve')

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Numerical Integration Result\nArea under the curve')

# Show the numerical result
ax.text(0, 70, f'Numerical Result: {result_numeric:.2f}', horizontalalignment='center', verticalalignment='center')

# Show the plot
plt.legend()
plt.show()

# Display the symbolic result
print(f"Symbolic Result: {result_symbolic}")
```
